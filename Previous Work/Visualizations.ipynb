```python
import seaborn as sns
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data from the CSV file
data = pd.read_csv('Data.csv')

```

```python
sns.catplot(data=data, x="Noise Multiplier", y="Testing Accuracy", kind="box", palette="Blues", height=6, aspect=1.4)
```

```python
sns.catplot(data=data, x="Noise Multiplier", y="MIA Accuracy", kind="box", palette="Blues", height=6, aspect=1.4)
```

```python
# Create a violin plot using Seaborn
plt.figure(figsize=(10, 6))

sns.violinplot(x="Noise Multiplier", y="Testing Accuracy", data=data,  palette="Blues", height=10, aspect=1.4)

# Display the plot
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.violinplot(x="Noise Multiplier", y="MIA Accuracy", data=data,  palette="Blues",  height=6, aspect=1.4)

# Display the plot
plt.show()
```
